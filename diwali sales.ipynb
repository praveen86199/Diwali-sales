```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
df
```

```python
df.info()
```

```python
df = df.drop(["Status", "unnamed1"], axis = 1)
df
```

```python
df.describe()
```

```python
df = df.dropna()
df["Amount"] = df["Amount"].astype(int)
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
df.describe()
```

```python
df["Marital_Status"] = df["Marital_Status"].astype(str)
```

```python
df["Marital_Status"] = df["Marital_Status"].replace("0", "Single")
df["Marital_Status"] = df["Marital_Status"].replace("1", "Married")
df
```

# Total no. of Males and Females

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df, palette = "viridis")
sns.set({"figure.figsize": (5,5)})
for x in ax.containers:
    ax.bar_label(x)
```

```python
df[['Age', 'Orders', 'Amount']].describe()
```

# Total no. of people in diff. age group

```python
ax = sns.countplot(x = "Age Group", data = df, hue = "Gender")
sns.set({"figure.figsize": (10,5)})

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),\
                ha='center', va='bottom', fontsize=8, \
                color='black', rotation=90, xytext=(0, 15), textcoords='offset points')
```

```python
df.head()
```

```python
gb = df.groupby("Product_Category").agg({"Amount" : "sum"})
gb = gb.sort_values("Amount", ascending = False)
gb["Amount in Million"] = gb["Amount"]/ 1000000
gb
```

# Comparison of total amount spent in diff. product category

```python
ax1 = sns.barplot(x = gb.index, y = gb["Amount in Million"], data = gb, palette = "viridis")
plt.xticks(rotation = 90)
sns.set({"figure.figsize": (20,5)})
for p in ax1.patches:
    ax1.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=10, color='red', rotation=90, xytext=(0, 15), textcoords='offset points')
```

```python
gb1 = df.groupby("Age Group").agg({"Amount" : "sum"})
gb1.sort_values("Amount", ascending = False)
gb1["Amount in Million"] = gb1["Amount"]/ 1000000
gb
```

# Comaprison of total amount spent by people of diff. Age group

```python
ax2 = sns.barplot(x = gb1.index, y = gb1["Amount in Million"], data = gb1, palette = "Pastel2")
sns.set({"figure.figsize": (10,5)})
for x in ax2.containers:
    ax2.bar_label(x)
```

# Comparision of total amount spent by diff. marital status

```python
sns.barplot(x = df["Marital_Status"], y = df["Amount"], estimator = "sum", data = df, hue = "Gender")
sns.set({"figure.figsize": (5,5)})
```

# Comparision of Amount spent in diff. states

```python
sns.set(rc = {'figure.figsize':(15,5)})
ax = sns.barplot(x = df["State"], y = df["Amount"], estimator = "sum", palette = "viridis", data = df, hue = "Gender")
plt.xticks(rotation = 90)

plt.show
```

# Total no. of buyers available in diff. zones

```python
ax = sns.countplot(x = df["Zone"], data = df, hue = "Gender", palette = "viridis")
sns.set({"figure.figsize": (10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# Comparision of Total Amount spent in diff. zones

```python
ax = sns.barplot(x = df["Zone"], y = df["Amount"], estimator = "sum", data = df, hue = "Gender", palette = "viridis")
sns.set({"figure.figsize": (20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.head()
```

# Comparison of Total Amount spent by diff. occupation

```python
sns.set({"figure.figsize" : (15,5)})
sns.barplot(x = df["Occupation"], y = df["Amount"], estimator = "sum", data = df, palette = "viridis")
plt.xticks(rotation = 90)
plt.show()
```
